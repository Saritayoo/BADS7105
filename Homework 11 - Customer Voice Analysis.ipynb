```python
# import pyLDAvis.gensim
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import pandas as pd
import pythainlp
import gensim
pyLDAvis.enable_notebook()
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

# Read file

```python
df = pd.read_csv('C:/Users/User/Desktop/CRM/Wongnai Reviews - Small.csv')
df.head()
```

```python
stop = list(pythainlp.corpus.thai_stopwords())
removed = [' ', '  ', '\n', 'ร้าน', '(', ')']
screening_words = stop + removed

def tokenize_with_space(sentence):
  merged = ''
  words = pythainlp.word_tokenize(str(sentence), engine='newmm')
  for word in words:
    if word not in screening_words:
      merged = merged + ',' + word
  return merged[1:]
df['Review_tokenized'] = df['Review'].apply(lambda x: tokenize_with_space(x))
df.head()
```

```python
list = df['Review_tokenized'].to_list()
texts = [[text for text in doc.split(',')] for doc in list]
dictionary = gensim.corpora.Dictionary(texts)
print(dictionary.token2id.keys())
```

```python
corpus = [dictionary.doc2bow(text, allow_update=True) for text in texts]
num_topics = 10
chunksize = 3000 # size of the doc looked at every pass
passes = 20 # number of passes through documents
iterations = 50
eval_every = 1  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

%time model = gensim.models.LdaModel(corpus=corpus, id2word=id2word, chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, num_topics=num_topics, \
                       passes=passes, eval_every=eval_every)
gensimvis.prepare(model, corpus, dictionary)
model.show_topic(1)
```

```python
df['topics'] = df['Review_tokenized'].apply(lambda x: model.get_document_topics(dictionary.doc2bow(x.split(',')))[0][0])
df['score'] = df['Review_tokenized'].apply(lambda x: model.get_document_topics(dictionary.doc2bow(x.split(',')))[0][1])
df.head()
```
